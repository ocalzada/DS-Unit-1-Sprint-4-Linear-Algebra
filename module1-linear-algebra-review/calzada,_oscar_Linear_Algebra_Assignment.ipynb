## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
C = np.array([[9,6,3],
             [4,7,11]]
            )
D = np.array([[1,0,0],
             [0,1,0],
             [0,0,1]]
            )

np.matmul(C,D)
# new CD matrix has a 2x3 dimension - 
# number of rows come from matrix C
# number of columns come from matrix D
```

```
A = np.array([[1,2],
             [3,4],
             [5,6]]
            )

E = np.array([[1,3],
             [5,7]]
            )

np.matmul(A,E)
# AE matrix has a 3x2 dimension - 
# number of rows comes from matrix A
# number of columns comes from matrix E
```

```
B = [2,4,6]
A = np.array([[1,2],
             [3,4],
             [5,6]]
            )

np.matmul(B, A)
# BA matrix has a 1x2 dimension - 
# number of rows comes from matrix B
# number of columns comes from matrix A
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
             [7,10]])

np.linalg.det(H)
#determinant of H
```

```
J = np.array([[0,1,2],
             [7,10,4],
             [3,2,0]])

np.linalg.det(J)
#determinant of J
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
#inverse of H
H_inv = np.linalg.inv(H)
print(H_inv)
```

```
#inverse of J
J_inv = np.linalg.inv(J)
print(J_inv)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
# H multiplied by its inverse

H_H_inv = np.matmul(H, H_inv)
print(H_H_inv)
```

```
# J multiplied by its inverse
J_J_inv = np.matmul(J, J_inv)
print(J_J_inv)
```

```
# the given identity matrices are not equivalent to each other because 
# their dimensions are not equivalent.
# the H identity matrix is a 2x2 matrix.
# the J identity matrix is a 3x3 matrix.
```
